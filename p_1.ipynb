```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=10)
```

```python
km.fit(x_transformed)
```

```python
km.labels_
```

```python
df
```

```python
df['group'] = km.labels_
```

```python
forth_group = df[df['group'] == 5]
forth_group['label'].value_counts()
```

```python
wcss = []
for i in range(5,50):
    km = KMeans(n_clusters=i)
    km.fit(x_transformed)
    wcss.append(km.inertia_)
```

```python
wcss
clusters = list(range(5,50))
clusters
```

```python
plt.plot(clusters,wcss)
plt.title("ELBOW METHOD")
plt.xlabel("clusters")
plt.ylabel("wcss") # wcss-> within a cluster standard square
plt.show()
```
